```python
from pylab import*
# Genera el vector x=[0, 0.1, 0.2, 0.3, . . . , 2]
x=arange(0,2,0.1) 
# Grafica los puntos (x, cos(x)) con una linea azul
plot(x,cos(x),'-b') 
# Grafica los puntos (x, log(x+1)) con una l√≠nea roja
plot(x,log(x+1),'-r') 
title('Funciones')
xlabel('x')
ylabel('f(x)')
grid(True)
legend(('coseno','logaritmo'),loc='upper right')
show()
```

```python
from pylab import*
from mpl_toolkits.mplot3d import Axes3D
fig=figure()
ax=Axes3D(fig)
x=arange(-4,4,0.25)
y=arange(-4,4,0.25)
X,Y=meshgrid(x,y)
Z=sin(sqrt(X**2+Y**2))
ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap='hot')
show()
```
